```python
import pandas as pd
```

```python
#Importing the dataset
df = pd.read_csv('https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv')
df
```

```python
#total rows & columns
print('Rows Count:',len(df))
print('Columns Count:',len(df.columns))
```

```python
# Data types of columns
print(df.dtypes)
```

```python
# Info & describe
print(df.info)
print(df.describe(include='all'))
```

```python
#unique values from columnn location
print('unique locations are: ',len(df['location'].unique()))
```

```python
# continent with maximum frequency 
name = df['continent'].value_counts().index.tolist()[0]
value = df['continent'].value_counts().tolist()[0]
print('continent has maximum frequency is',name,' and count is',value)
```

```python
# max & mean values of 'total_cases' column
print('max value:',df['total_cases'].max())
print('mean value:',df['total_cases'].mean())
```

```python
# 25%,50% & 75% quartile value of 'total_deaths' column
df['total_deaths'].describe()
```

```python
#continent having max human_development_index
maxvalue = df['human_development_index'].max()
df[df['human_development_index'] == maxvalue]['continent'].unique()
```

```python
#continent which has minimum gdp_per_capita
min_value = df['gdp_per_capita'].min()
df[df['gdp_per_capita'] == min_value]['continent'].unique()
```

```python
# Filtering dataframe and remove duplicates
filtered_df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
filtered_df = filtered_df.drop_duplicates()
```

```python
#missing values
filtered_df.isnull().sum()
```

```python
#Removing missing and fill zero
filtered_df = filtered_df[filtered_df['continent'].isnull()==False]
filtered_df = filtered_df.fillna(0)
```

```python
#Creating month column 
filtered_df['month'] = pd.DatetimeIndex(filtered_df['date']).month
filtered_df
```

```python
# max value of all 'continent' values and resetting the index
df_groupby = filtered_df.groupby('continent').max()
df_groupby.reset_index(inplace = True)
```

```python
#Creating a new feature 'total_deaths_to_total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/ df_groupby['total_cases']
df_groupby
```

```python
import seaborn as sns
```

```python
#gdp_per_capita histogram using seaborn dist plot.
sns.distplot(df_groupby['gdp_per_capita'])
```

```python
# scatter plot of 'total_cases' with 'gdp_per_capita'
sns.scatterplot(data = df_groupby,x='total_cases',y='gdp_per_capita')
```

```python
# Pairplot of df_groupby dataframe
sns.pairplot(df_groupby)
```

```python
#bar plot of 'continent' with 'total_cases'
sns.catplot(data=df_groupby, x='continent',y='total_cases', kind='bar')
```
